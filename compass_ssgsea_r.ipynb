```r
library(GSVA)

library(reticulate)
```

```r
out_file_path <- './compass_comparison/scrna/'
```

# read in DE genes

```r
de_genes <- read.csv(paste(out_file_path, 'de_genes_mann_whitney.csv', sep = ''))
de_genes
```

```r
library(dplyr)
library(tidyr)
```

```r
de_genes <- read.table(paste(out_file_path, 'de_genes_format.csv', sep = ''), sep = ',', header = TRUE, row.names = 1)
de_genes
```

```r
v1 <- unlist(de_genes[,1])
v2 <- unlist(de_genes[,2])

df <- matrix(c(v1,v2), nrow = 795, ncol = 2)
rownames(df) <- rownames(de_genes)
colnames(df) <- colnames(de_genes)
df
```

# read in subsystems

```r
subsystems <- read.csv(paste(out_file_path, 'subsystems_filtered.csv', sep = ''))
subsystems
```

```r
cell_d_list <- apply(subsystems, 2, as.list)
cell_d_list$X <- NULL
cell_d_list <- lapply(cell_d_list, function(z){ z[!is.na(z) & z != ""]})
```

```r
de_genes
```

```r
df[1,]
```

```r
gsva_es <- gsva(df, cell_d_list, method = 'ssgsea', abs.ranking = FALSE, ssgsea.norm = TRUE)
```

```r
gsva_es
```

```r
write.csv(gsva_es, file = paste(out_file_path, 'ssgea_res.csv', sep = ''), sep = ',')
```
